```python
small_df = df.drop(columns = ['회원번호', '남녀구분코드', '연령', '자격구분상세', '거주시도명', '직장시도명', 'Life_Stage', '연령대', '이용금액_납부', '이용금액_쇼핑', '이용금액_교통', '이용금액_교육', '이용금액_여유생활', '쇼핑_기타_이용금액', '여유_기타이용금액', '교통_주유이용금액', '교통_정비이용금액',
       '교통_통행료이용금액', '이용금액_일상생활'])
```

```python
small_df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_df = pd.DataFrame(scaler.fit_transform(small_df), columns=small_df.columns)
```

```python
scaled_df = scaled_df * 10000
```

```python
scaled_df.columns
```

```python
dict_list = []

# 각 행에 대해 딕셔너리를 생성하고 리스트에 추가
for idx, row in small_df.iterrows():
    row_dict = {col: value for col, value in row.items()}
    dict_list.append(row_dict)
```

```python
dict_list
```

```python
import pandas as pd
import gensim
import gensim.corpora as corpora
from gensim.models.ldamodel import LdaModel
import pyLDAvis
import pyLDAvis.gensim_models
import json
```

```python
texts = [[str(k) for k, v in doc.items() if v > 0] for doc in dict_list]
id2word = corpora.Dictionary(texts)
corpus = [id2word.doc2bow(text) for text in texts]
```

```python
# LDA 모델 학습
# LDA 모델 학습
lda_model = LdaModel(corpus=corpus,
                     id2word=id2word,
                     num_topics=10,
                     random_state=100,
                     update_every=1,
                     chunksize=20,
                     passes=10,
                     alpha='symmetric',  # alpha 값을 symmetric으로 설정
                     eta='auto',
                     per_word_topics=True)
```

```python
# LDA 시각화
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word)
pyLDAvis.display(vis)
```

```python
for i, doc in enumerate(corpus):
    topics = lda_model.get_document_topics(doc)
    dominant_topic = max(topics, key=lambda x: x[1])[0]  # 가장 높은 확률의 주제 선택
    scaled_df.loc[i, 'label'] = dominant_topic + 1  # 주제 번호를 1부터 시작하도록 인덱스 조정
```

```python
r1 = [x for x in df.columns if "이용금액_" in x]

new_data = pd.DataFrame()

for a in range(1, 10):
    condz = dfz["label"] == a
    if condz.any():  # 해당 주제에 속하는 데이터가 있는 경우에만 처리
        new_data = pd.DataFrame()
        for b in r1:
            ak = dfz.loc[condz, b].mean()
            new_data.loc[0, b] = ak  # 새로운 데이터프레임에 평균값 저장
        globals()[f"new_data{a}"] = new_data  # 변수명으로 저장
        print(f"new_data{a}:\n{new_data}\n")
        
```

```python
import plotly.express as px

for i in range(1, 10):
    # 'new_data1', 'new_data2', ..., 'new_data9'와 같은 변수에 접근
    df_melted = eval(f'new_data{i}').melt(var_name='Category', value_name='Amount')

    # 각 카테고리에 대해 히스토그램 생성
    fig = px.histogram(df_melted, x='Category', y='Amount', title=f'이용금액 분포 {i}')
    fig.show()
```
